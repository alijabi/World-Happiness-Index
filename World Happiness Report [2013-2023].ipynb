```python
df_2013 = df[df['Year'] == 2013]
df_2013_sorted = df_2013.sort_values('Rank', ascending=True)

# data for 2014 is missing
#df_2014 = df[df['Year'] == 2014]
#df_2014_sorted = df_2014.sort_values('Rank', ascending=True)

df_2015 = df[df['Year'] == 2015]
df_2015_sorted = df_2015.sort_values('Rank', ascending=True)

df_2016 = df[df['Year'] == 2016]
df_2016_sorted = df_2016.sort_values('Rank', ascending=True)

df_2017 = df[df['Year'] == 2017]
df_2017_sorted = df_2017.sort_values('Rank', ascending=True)

df_2018 = df[df['Year'] == 2018]
df_2018_sorted = df_2018.sort_values('Rank', ascending=True)

df_2019 = df[df['Year'] == 2019]
df_2019_sorted = df_2019.sort_values('Rank', ascending=True)

df_2020 = df[df['Year'] == 2020]
df_2020_sorted = df_2020.sort_values('Rank', ascending=True)

df_2021 = df[df['Year'] == 2021]
df_2021_sorted = df_2021.sort_values('Rank', ascending=True)

df_2022 = df[df['Year'] == 2022]
df_2022_sorted = df_2022.sort_values('Rank', ascending=True)

df_2023 = df[df['Year'] == 2023]
df_2023_sorted = df_2023.sort_values('Rank', ascending=True)

df_2020_sorted
```

## Mean World Happiness Index [2013-2023]

```python
dfs = [df_2013_sorted, df_2015_sorted, df_2016_sorted, 
       df_2017_sorted, df_2018_sorted, df_2019_sorted, df_2020_sorted, 
       df_2021_sorted, df_2022_sorted, df_2023_sorted]
years = [2013, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

dfs_mean = []

for data in dfs:
    dfs_mean.append(round(data['Index'].mean(), 3))

print(dfs_mean)

x = years
y = dfs_mean

fig, ax = plt.subplots()
plt.yticks(np.arange(5.2, 6.0, step=.05))
plt.ylabel('Mean World Happiness Index')
plt.xlabel('Years')
plt.xticks(years)
plt.title('Mean World Happiness Index [2013-2023]')
plt.plot(x, y, marker='o')
plt.show()
```

## Happiest vs. Unhappiest Countries [2013-2024]

```python
dfs = [df_2013_sorted, df_2015_sorted, df_2016_sorted, 
       df_2017_sorted, df_2018_sorted, df_2019_sorted, df_2020_sorted, 
       df_2021_sorted, df_2022_sorted, df_2023_sorted]
years = [2013, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

top_countries = []

for data in dfs:
    top_countries.append(data['Country'].iloc[0])
    
dfs_top_index = []

for data in dfs:
    dfs_top_index.append(data['Index'].iloc[0])

bottom_countries = []

for data in dfs:
    bottom_countries.append(data['Country'].iloc[-1])
    
dfs_bottom_index = []

for data in dfs:
    dfs_bottom_index.append(data['Index'].iloc[-1])
    

print(top_countries)
print(dfs_top_index)
print(bottom_countries)
print(dfs_bottom_index)

x = years
y = dfs_top_index
y2 = dfs_bottom_index

fig, ax = plt.subplots(figsize=(10,6))
plt.yticks(np.arange(7.2, 8.0, step=.05))
plt.ylabel('World Happiness Index')
plt.xlabel('Years')
plt.xticks(years)
plt.title('Happiest vs. Unhappiest Countries [2013-2023]')

# merge lists to tupel
merged_list = [(x[i], y[i]) for i in range(0, len(x))]
merged_list2 = [(x[i], y2[i]) for i in range(0, len(x))]


# Add labels to top countries on the line
label_points = merged_list  # Example points to label
label_texts = top_countries  # Corresponding labels

for point, label in zip(label_points, label_texts):
    plt.text(point[0], point[1], label, ha='right', va='bottom')

ax2 = ax.twinx()    

# Add labels to bottom countries on the line
label_points2 = merged_list2  # Example points to label
label_texts2 = bottom_countries  # Corresponding labels

for point, label in zip(label_points2, label_texts2):
    plt.text(point[0], point[1], label, ha='right', va='bottom')
    

ax.plot(x, y, marker='o', color='green')
ax2.plot(x, y2, marker='o', color='red')
plt.show()
```
